# Pandas Filtering Or Subsetting a Dataframe

```
#@title Select subset of A dataframe and return a Dataframe with Boolean values
```

```
import numpy as np
import pandas as pd
```

```
randomnum=np.random.randint(-100,100,20)
randomnum
```

```
randomnum2d=randomnum.reshape(5,4)
randomnum2d
```

```
rowindex=["A","B","C","D","E"]
colindex=["W","X","Y","Z"]
exampleDF=pd.DataFrame(randomnum2d,rowindex,colindex)
exampleDF
```

```
exampleDF>0
```

```
#@title Select subset of A dataframe and filter the same instead of returning boolean values
exampleDF[exampleDF>0]
```

```
exampleDF
```

```
#@title  Return a Series from dataframe with Boolean values based on a condition for a column in dataframe
exampleDF['X']>0
```

```
#@title  Return a dataframe with actual instead of boolean values based on a condition for a column in dataframe
exampleDF[exampleDF['X']>0]
```

```
#@title  Return a series with actual instead of boolean values based on a condition for a column in dataframe and then projecting a particular column only
exampleDF[exampleDF['X']>0]['Y']
```

```
#@title  Return a series with actual instead of boolean values based on a condition for a column in dataframe and then projecting a particular row using names

exampleDF[exampleDF['X']>0].loc[['C','E']]
```

```
#@title  Return a series with actual instead of boolean values based on a condition for a column in dataframe and then projecting a particular row using index

exampleDF[exampleDF['X']>0].iloc[['0','2']]
```

```
#@title  Multiple Conditions using And/Or operator
exampleDF
```

# Pandas GroupBy

```
import numpy as np
import pandas as pd
```

```
empsales=pd.DataFrame([["Jassi",450],["Pradeep",100],["Yogesh",300],["Jassi",200],["Pradeep",600],["Yogesh",900]],columns=["Name","Sales"])
empsales
```

```
empsales.head()
```

```
#@title  Group by then Sum

empsales.groupby(by='Name').sum()
```

```
#@title  Group by then Mean

empsales.groupby(by='Name').mean()
```

```
#@title  Sort rows by index after grouping
empsales.groupby(by='Name').mean().sort_index()
```

```
empsales.groupby(by='Name').mean().sort_index(ascending=False)
```

```
#@title  Group by Multiple Columns
empyrsales=pd.DataFrame([["Jassi","2010",450],["Pradeep","2012",100],["Yogesh","2010",300],["Jassi","2010",200],["Pradeep","2016",600],["Yogesh","2014",900]],columns=["Name","Year","Sales"])
empyrsales
```

```
empyrsales.groupby(by=["Name","Year"]).sum()
```

```
#@title  Combine Group by with describe for useful statistics
empyrsales.groupby(by=["Name"]).describe()
```

```
empyrsales.groupby(by=["Name"]).describe().transpose()
```
