# Matplotlib

```
#@title Functional Matplotlib example
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
x=np.linspace(100,500,11)
y=x**2
```

```
#@title Add X and Y Label and Title
```

```
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Sample Plot")
```

```
#@title Multiplot/Subplot on same canvas
```

```
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
#@title Object Oriented Matplotlib example
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Sample Plot")
```

```
#@title Object Oriented Matplotlib example with 2 figures
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
```

```
#@title Multiplot/Subplot Using Object oriented Matplotlib
#Subplot function creates axis for you based on number of rows and column
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```
fig,axes=plt.subplots(nrows=3,ncols=3)
```

```
#@title Use tight layout to avoid overlapping
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
```

```
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].set_title("FirstPlot")
axes[1].set_title("SecondPlot")
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```
#@title Legends for multiple plots in the same axes
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**3)
axes.set_title("Multiple Plots")
```

```
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="Twice")
axes.plot(x,x**3,label="Thrice")
axes.set_title("Multiple Plots")
axes.legend()
```

```
#Changing positions of legend(https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.legend.html)
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="Twice")
axes.plot(x,x**3,label="Thrice")
axes.set_title("Multiple Plots")
axes.legend(loc=9)
```

```
#USing tuple just like for axes to specify legends
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="Twice")
axes.plot(x,x**3,label="Thrice")
axes.set_title("Multiple Plots")
axes.legend(loc=(0.5,0.1))
```

# Seaborne

```
#@title 1.Distribution Plots
```

```
#load built in tips dataset
import seaborn as sns
tips=sns.load_dataset("tips")
```

```
tips.head(10)
```

```
#Histogram to show distribution of where your total_bill lies(Y axis has counts X has total bill)
sns.distplot(tips["total_bill"],kde=False)
```

```
tips.describe()
```

```
tips[(tips["total_bill"]>=10) & (tips["total_bill"]<=20)].describe()
```

```
#@title Changing number of Bins for a distribution plot
sns.distplot(tips["total_bill"],kde=False,bins=30)
```

```
sns.distplot(tips["total_bill"],kde=False,bins=100)
```

```
sns.distplot(tips["total_bill"],kde=False,bins=3)
```

```
#@title Joint Plot -> to combine two distribution plots(bi variate data or 2 variables)
```

```
#shows how tips is propotional to total_bill
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```
#@title kind parameter to change what is happening inside jointplot,
#Hexagon gets darker if it has more number of points instead of having all scatter points on
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex")
```

```
sns.jointplot(x="total_bill",y="tip",data=tips,kind="regression")
```

```
#@title This shows a 2 dimensional KDE which effectively shows you density of where these points match up the most 
sns.jointplot(x="total_bill",y="tip",data=tips,kind="kde")
```

```
#@title Pairpplot -> is going to plot pairwise relationships across an entire dataframe atleast numerical columns
# It is going to do a joinplot for every single combination of numerical colums in this dataframe
#column vs column is a histogram instead because a scatter plot would show straight line
sns.pairplot(tips)
```

```
#@title Use hue argument in Pairplots to mark the categorical colums in the data points ex:Sex

sns.pairplot(tips,hue="sex")
```

```
#@title Rugplot(Out of scope)-> In contrast to Distributon plot you just make a dashmark for every single datapoint
sns.rugplot(tips["total_bill"])
```

```
#@title Kernel Density estimation(Out of scope)
sns.distplot(tips["total_bill"])
```

```
#@title 2.Categorical Plots
import seaborn as sns
import numpy as np
tips=sns.load_dataset("tips")
```

```
tips.head()
```

```
#@title BarPlot -> It shows you the mean/average of total bill for each categorical value 
sns.barplot(x="sex",y="total_bill",data=tips)
```

```
#@title Change estimator from mean to something else for barplot
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std)
```

```
#@title CountPlot -> To count number of occurences of categorical value
sns.countplot(x="sex",data=tips)
```

```
tips["sex"].describe()
```

```
#@title Boxplot ->  box plot or boxplot is a method for graphically depicting groups of numerical data through their quartiles. Box plots may also have lines extending from the boxes indicating variability outside the upper and lower quartiles, hence the terms box-and-whisker plot and box-and-whisker diagram.
sns.boxplot(x="day",y="total_bill",data=tips)
```

```
# Total Bill Quartile for entire dataset
tips.describe()
```

```
# Total Bill Quartile for entire dataset based on each day
tips.groupby("day").describe()
```

```
#@title Use hue with Boxplot to show another different boxplot for different categorical column 

sns.boxplot(x="day",y="total_bill",data=tips,hue="sex")
```

```
#@title stripplot ->  scatterplot based of the category
sns.stripplot(x="day",y="total_bill",data=tips)
```

```
#@title Split stripplot on another categorical column using hue
sns.stripplot(x="day",y="total_bill",data=tips,hue="sex",split=True,jitter=True)
```

```
#Verify the above data
tips[tips["sex"]=="Male"].groupby("day").describe()
```

```
tips[tips["sex"]=="Female"].groupby("day").describe()
```

```
import seaborn as sns
import numpy as np
tips=sns.load_dataset("tips")
flights=sns.load_dataset("flights")
```

```
tips.head()
```

```
flights.head()
```

```
#@title A heat map -> is a graphical representation of data where the individual values contained in a matrix are represented as colors. "Heat map" is a newer term but shading matrices have existed for
```

```
# Pivot flights table based on year
flight_pivot=flights.pivot(index="month",columns="year",values="passengers")
```

```
flight_pivot
```

```
sns.heatmap(flight_pivot)
```

```
#@title Use cmap with heatmap to change color schme
sns.heatmap(flight_pivot,cmap="coolwarm")
```

```
#@title Use linecolor and linewidth to apply some separation
sns.heatmap(flight_pivot,cmap="coolwarm",linewidth=1,linecolor="black")
```

```
#@title A cluster map -> will cluster things an order such that similar groups are close to each other
sns.clustermap(flight_pivot,cmap="coolwarm",linewidth=1,linecolor="black")
```
